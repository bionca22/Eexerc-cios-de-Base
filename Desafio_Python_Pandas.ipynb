```python
import pandas as pd
```

```python
d2women=data("women")
```

```python
d2women.head()
```

```python
pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

### 1. Construa uma nova representação onde os nomes das colunas estão em português, e os valores de altura são convertidos para centímetros e os valores de peso convertidos para quilograma.

```python
d2women = data("women")

polegada_para_cm = 2.54
libra_para_kg = 0.45

nova_representacao = pd.DataFrame({
    'altura': d2women['height'] * polegada_para_cm,
    'peso': d2women['weight'] * libra_para_kg
})

print(nova_representacao)
```

### 2. Plote o scatterplot de altura x peso da nova representação.

```python
nova_representacao.plot.scatter(x='altura', y='peso')
plt.title('altura x peso')
plt.xlabel('altura (cm)')
plt.ylabel('peso (kg)')
plt.show()
```

### 2. Quantas pessoas eram da primeira, segunda e terceira classe?

### 3. Qual a porcentagem de sobreviventes na primeira classe?

```python
sobreviventes[-2:]
```

#Desafio 5: Desafio 5: Forbes - Gravar Vídeo

dupla: Todo Squad <3

```python
from pydataset import data
```

```python
df_forbes=data("Forbes2000")
df_forbes
```

## 1. Qual a empresa mais valiosa da lista da Forbes 2000?

```python
mais_valioso = df_forbes.sort_values('marketvalue', ascending = False).head(1)
mais_valioso
```

## 2. Construa uma tabela com as TOP 10 empresas mais lucrativas da lista.

```python
top_10_empresas_lucrativas = df_forbes.nlargest(10, 'profits')
tabela_nome_lucro = top_10_empresas_lucrativas[['name', 'profits']]
print(tabela_nome_lucro)
```

## 3. Qual a média de valores de empresa das cinco categorias mais valiosas?

```python
top_5_empresas_mais_valiosas = df_forbes.groupby("category")['marketvalue'].mean().head()
top_5_empresas_mais_valiosas
```
